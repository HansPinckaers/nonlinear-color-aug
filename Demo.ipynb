```python
import PIL
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

from aug import NonLinearColorAugmentation 
```

```python
augment_img = NonLinearColorAugmentation()
```

```python
img = PIL.Image.open('glands.png')
img
```

```python
fig, ax = plt.subplots(figsize=(15, 10.3), ncols=3, nrows=3)

np_img = np.array(img)

for row in range(3):
    for col in range(3):
        ax[row, col].imshow(augment_img(np_img))
        ax[row, col].set_axis_off()

fig.tight_layout()
```
